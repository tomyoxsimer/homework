```python
# Assingment 7 Thomas Yoxsimer

import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

def sort_tuple(alist):
    alist.sort(key = lambda x: x[1])
    return alist

G = nx.read_edgelist('stack_overflow_edgelist.csv',delimiter=',', create_using=nx.Graph()) # read in the data and create networkX graph 

print(nx.info(G)) # display the totals for nodes and edges

nx.draw(G, with_labels=True, font_weight='bold') # draw and show the networkX graph
plt.show(G)

pr = nx.pagerank(G) # Pageranks everything
pr_list = []

for x in pr: # Turns dictionary into list of tuples
    pr_list.append((x, pr[x]))

del pr_list [0:2] # remove 'source' and 'target'

sort_tuple(pr_list) # sorts tuple using function above 

print('Bottom Ten')
for x in range(10):
    print(pr_list[x])
    
pr_list.reverse()

print('\nTop Ten')
for x in range(10):
    print(pr_list[x])
```
